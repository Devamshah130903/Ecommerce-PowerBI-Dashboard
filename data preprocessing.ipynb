```python
# Fill missing text values with 'Unknown'
df=df.fillna({
    'country': 'Unknown',
    'category': 'Unknown',
    'payment_method': 'Unknown'
})

# If any numeric column has missing values
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    df[col].fillna(df[col].median(), inplace=True)
```

# Fix Data Types

```python
# Convert date columns
date_cols = ['signup_date', 'order_date', 'review_date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# Convert category-type columns to 'category'
cat_cols = ['gender', 'age_group', 'country', 'category', 'payment_method', 'order_status']
for col in cat_cols:
    df[col] = df[col].astype('category')
```

# Feature Engineering

```python
df['Total_Sales'] = df['quantity'] * df['unit_price']
```

```python
df['Profit'] = df['Total_Sales'] * 0.2
```

```python
df['Profit_Margin_%'] = (df['Profit'] / df['Total_Sales']) * 100
```

```python
df['Order_Year'] = df['order_date'].dt.year
df['Order_Month'] = df['order_date'].dt.month_name()
df['Order_Day'] = df['order_date'].dt.day
```

```python
df['Customer_Name'] = df['first_name'] + ' ' + df['last_name']
```

# Remove Invalid or Extreme Values

```python
# Remove rows where quantity or price <= 0
df = df[(df['quantity'] > 0) & (df['unit_price'] > 0)]

# Cap extreme Total_Sales (optional)
df = df[df['Total_Sales'] < df['Total_Sales'].quantile(0.99)]
```

# Clean Category Values

```python
df['category'] = df['category'].str.title().str.strip()
df['country'] = df['country'].str.title().str.strip()
```

# Save Dataset

```python
# Save to new CSV
df.to_csv("dataset/ecommerce_cleaned.csv", index=False)

print("✅ Data cleaning complete! Cleaned file saved as 'ecommerce_cleaned.csv'")
```
